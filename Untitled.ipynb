```python
from Twitterbot_files.DatabaseUtils import *
import unicodedata
import re
```

```python
db_connection = DatabaseConnector("localhost", "root", "12345", "twitter_database")
```

```python
#Recuperar todos os tweets de notícias

sql = "SELECT content FROM newswire_tweets"
results = db_connection.selectSQLCommand(sql)
```

```python
# Corrigindo os erros de encoding dos tweets de notícias
tweets = [re.sub("(https://.+)", "", tweet[0]).strip() for tweet in results]
tweets = [tweet.encode(encoding="cp860", errors = "ignore").decode("cp860") for tweet in tweets]
```

```python
from nltk import FreqDist, word_tokenize
from nltk.corpus import stopwords

# word_tokenize(tweets[0])
freqdist = FreqDist([word.lower() for tweet in tweets for word in word_tokenize(tweet) if word not in stopwords.words("portuguese")+[".", ",", ":", "'", '"', "?", "!", "@", "''", "``"]])
```

```python
freqdist.most_common()
```
